```python
import pandas as pd
import numpy as np

df = np.genfromtxt("/Users/saibalajikondapalli/Documents/Python Practice/Student-Performance-UCI/student-mat.csv", delimiter=';', dtype = None)
print(df)
```

```python
df.columns = [name.decode('utf8') for name in df.iloc[0]]

df = df.drop(df.index[0])

df.reset_index(drop=True, inplace=True)

df.head()
```

```python
def clean_data(value):
    if isinstance(value, bytes):
        return value.decode('UTF-8').replace('"','')
    return value

df = df.applymap(clean_data)
```

```python
df.head()
```

```python
pd.set_option('display.max_rows', 100)
df.nunique()
```

```python
df.info()
```

```python
df = df.convert_dtypes()

num_cols = ['age','Medu','Fedu','traveltime','studytime','failures','famrel','freetime','goout','Dalc','Walc','health','absences','G1','G2','G3']

df[num_cols] = df[num_cols].astype(int)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df = df.drop('reason', axis = 1)

str_cols = ['school','sex','address','famsize','Pstatus','Medu','Fedu','Mjob','Fjob','guardian','schoolsup','famsup','paid','activities','nursery','higher','internet','romantic']

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df[str_cols] = df[str_cols].apply(lambda col: le.fit_transform(col))
```

```python
from sklearn.preprocessing import StandardScaler

ss_cols = ['age', 'absences', 'G1', 'G2', 'G3']
ss = StandardScaler()
df[ss_cols] = ss.fit_transform(df[ss_cols])
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split

x = df.drop('G3', axis=1)
y = df['G3']

X_train, X_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train, y_train)

y_pred = lr.predict(X_test)

from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_pred, y_test)
mse = mean_squared_error(y_pred, y_test)

print(f'The R2 score is: {r2}\n')
print(f'The Mean average score is: {mae}\n')
print(f'The Mean squared error is: {mse}')
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Lasso


alphas = [0.01,0.1,0.2,0.5, 1,10]
l = Lasso()

grid_search = GridSearchCV(estimator=l, param_grid={'alpha':alphas}, scoring='neg_mean_squared_error', cv=10)
grid_search.fit(X_train,y_train)

best_alpha = grid_search.best_params_['alpha']

ls = Lasso(alpha = best_alpha)
ls.fit(X_train, y_train)

y_pred_ls = ls.predict(X_test)

r2 = r2_score(y_test, y_pred_ls)
mae = mean_absolute_error(y_pred_ls, y_test)
mse = mean_squared_error(y_pred_ls, y_test)

print(f'The R2 score is: {r2}\n')
print(f'The Mean average score is: {mae}\n')
print(f'The Mean squared error is: {mse}')
```

```python
from sklearn.tree import DecisionTreeRegressor

dt = DecisionTreeRegressor(criterion='absolute_error', random_state=42)

dt.fit(X_train, y_train)
y_pred_dt = dt.predict(X_test)


r2 = r2_score(y_test, y_pred_dt)
mae = mean_absolute_error(y_pred_dt, y_test)
mse = mean_squared_error(y_pred_dt, y_test)

print(f'The R2 score is: {r2}\n')
print(f'The Mean average score is: {mae}\n')
print(f'The Mean squared error is: {mse}')
```

```python
from sklearn.ensemble import RandomForestRegressor

r = RandomForestRegressor()

param_grid = {'n_estimators':[100,200,300],
              'max_depth': [None, 10,20,30],
              'min_samples_split': [2,5,10],
              'min_samples_leaf':[1,2,4],
              'max_features': ['sqrt','log2']
              }
grid_search_rf = GridSearchCV(estimator=r, param_grid=param_grid, cv=5, scoring='neg_mean_squared_error', verbose=2, n_jobs=-1)
grid_search_rf.fit(X_train, y_train)

print('The best parameters are:', grid_search_rf.best_estimator_)
best_model = grid_search_rf.best_estimator_

y_pred_rf = best_model.predict(X_test)
r2 = r2_score(y_test, y_pred_rf)
mae = mean_absolute_error(y_pred_rf, y_test)
mse = mean_squared_error(y_pred_rf, y_test)

print(f'The R2 score is: {r2}\n')
print(f'The Mean average score is: {mae}\n')
print(f'The Mean squared error is: {mse}')
```

```python
#Random Forest Without Grid Search
rd = RandomForestRegressor(n_estimators=100)

rd.fit(X_train,y_train)
yj = rd.predict(X_test)

r2 = r2_score(y_test, yj)
mae = mean_absolute_error(yj, y_test)
mse = mean_squared_error(yj, y_test)

print(f'The R2 score is: {r2}\n')
print(f'The Mean average score is: {mae}\n')
print(f'The Mean squared error is: {mse}')
```

```python
from sklearn.svm import SVR

sv_r = SVR(kernel='linear')

sv_r.fit(X_train, y_train)
y_pred_sv_r = sv_r.predict(X_test)


r2 = r2_score(y_test, y_pred_sv_r)
mae = mean_absolute_error(y_pred_sv_r, y_test)
mse = mean_squared_error(y_pred_sv_r, y_test)

print(f'The R2 score is: {r2}\n')
print(f'The Mean average score is: {mae}\n')
print(f'The Mean squared error is: {mse}')
```

```python
sv = SVR()

param_grid_svr = {'kernel':['linear','rbf','poly','sigmoid'],
                  'C':[0.1,1,10],
                  'epsilon':[0.1,0.2,0.5],
                  'gamma':['scale','auto'],
                  'degree':[2,3,4]
                  }
grid_search_svr = GridSearchCV(estimator=sv, param_grid=param_grid_svr, scoring='neg_mean_squared_error', cv=5)
grid_search_svr.fit(X_train,y_train)

best_params_svr = grid_search_svr.best_params_
best_model_svr = grid_search_svr.best_estimator_

y_pred_svr = best_model_svr.predict(X_test)


r2 = r2_score(y_test, y_pred_svr)
mae = mean_absolute_error(y_pred_svr, y_test)
mse = mean_squared_error(y_pred_svr, y_test)

print(f'The R2 score is: {r2}\n')
print(f'The Mean average score is: {mae}\n')
print(f'The Mean squared error is: {mse}')
```

```python
from sklearn.ensemble import GradientBoostingRegressor

gb_r = GradientBoostingRegressor()
gb_r.fit(X_train,y_train)

y_pred_gb_r = gb_r.predict(X_test)
r2 = r2_score(y_test, y_pred_gb_r)
mae = mean_absolute_error(y_pred_gb_r, y_test)
mse = mean_squared_error(y_pred_gb_r, y_test)

print(f'The R2 score is: {r2}\n')
print(f'The Mean average score is: {mae}\n')
print(f'The Mean squared error is: {mse}')
```

```python
gb = GradientBoostingRegressor()

param_grid_gb = {'loss':['squared_error', 'absolute_error', 'huber', 'quantile'],
                 'learning_rate':[0.01,0.05,0.1,0.2],
                 'n_estimators':[50,100,200],
                 'max_depth':[3,4,5],
                 'min_samples_split':[2,5,10],
                 'min_samples_leaf':[1,2,4],
                 'max_features':[None,'sqrt','log2',0.5,0.75]
                 }
grid_search_gb = GridSearchCV(estimator=gb, param_grid=param_grid_gb, scoring='neg_mean_squared_error', cv = 5)
grid_search_gb.fit(X_train, y_train)

best_params_gb = grid_search_gb.best_params_

best_model_gb = GradientBoostingRegressor(**best_params_gb)
best_model_gb.fit(X_train,y_train)

y_pred_gb = best_model_gb.predict(X_test)

y_pred_gb = gb_r.predict(X_test)
r2 = r2_score(y_test, y_pred_gb)
mae = mean_absolute_error(y_pred_gb, y_test)
mse = mean_squared_error(y_pred_gb, y_test)

print(f'The R2 score is: {r2}\n')
print(f'The Mean average score is: {mae}\n')
print(f'The Mean squared error is: {mse}')
```
